```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Criando nosso DataFrame
df = pd.read_excel("./Datasets/AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total
df["Valor Venda"].sum()
```

```python
#Qual o custo total
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Custo Total
round(df["custo"].sum(),2)
```

```python
#Lucro Total
df["lucro"] = df["Valor Venda"] - df['custo']
```

```python
df.head(1)
```

**A média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna "tempo_envio" em numéricos**

```python
#Extraindo apenas os dias
df["tempo_envio"] = (df["Data Envio"] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo de coluna "tempo_envio"
df["tempo_envio"].dtype
```

```python
#Média do tempo de envio por marca
df.groupby("Marca")["tempo_envio"].mean()
```

**Missing Values**

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

**Lucro por Ano e por Marca** 

```python
#Agrupar por ano e por data
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#Formatar dados
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Colocar index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos ordenado do maior pro menor?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).plot.barh(title="Total produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Gráfico Total de produtos vendidos
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Gráfico de Linhas
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
# Grafico de Barras - por marca
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal'); #info na horizontal
```

```python
#Gráfico de Barras - por classe
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
